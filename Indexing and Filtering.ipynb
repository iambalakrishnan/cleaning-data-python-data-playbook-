<h4> Filtering Data with str.contains

```python
data = pd.read_csv('artwork_sample.csv')
data.head()
```

```python
data['medium']
```

```python
type(data['medium'])
```

```python
data['medium'].str
```

```python
data[data['medium'].str.contains('Graphite')][['id','title','medium']]
```

```python
data.loc[data['medium'].str.contains('Graphite'),['id','title','medium']]
```

```python
#str.contains method is case sensitive we can change that by passing cas=False
data.loc[data['medium'].str.contains('Graphite', case=False),['id','title','medium']]
```

```python
#we can use regular expression by passing regex method

data.loc[data['medium'].str.contains('(?i)Graphit', regex=True),['id','title','medium']]
```

```python
data.loc[data['medium'].str.contains('(?i)Graphite', regex=True) | data['medium'].str.contains('(?i)Line', regex=True),['id','title','medium']]
```

```python
data.loc[data['medium'].str.contains('Graphite', case=False) |data['medium'].str.contains('Line', case=False),['id','title','medium']]
```

```python
#Using regex method
data.loc[data['medium'].str.contains('(?i)Graphite|Line', case=False, regex=True),['id','artist','medium','title']]
```

```python
data.loc[data['medium'].str.contains('Graphite|Line', case=False),['id','artist','medium']]
```

```python
data['year'].astype(str).str.contains('1826')
```

```python
#Above method willnot chanhe the original data frame
data.dtypes
```

```python
#We can apply  same operation using loc method
data.loc[data['year'].astype(str).str.contains('1826')].head(3)
```
